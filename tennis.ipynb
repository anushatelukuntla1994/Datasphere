```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn
```

```python
df = pd.read_csv('play_tennis.csv')
```

```python
df

##objective: to decide if we can play tennis or not
```

```python
## data prep and cleaning
```

```python
df_new = df.copy()
```

```python
# categorical encoding
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
## initiate a label encoder object
outlook_le = LabelEncoder()
```

```python
outlook_le.fit_transform(df_new['outlook'])
```

```python
df_new['outlook']   = outlook_le.fit_transform(df_new['outlook'])
```

```python
df_new
```

```python
temp_le = LabelEncoder()
humidity_le = LabelEncoder()
wind_le = LabelEncoder()
play_le = LabelEncoder()
```

```python
df_new['temp']=temp_le.fit_transform(df_new['temp'])
df_new['humidity']=humidity_le.fit_transform(df_new['humidity'])
df_new['wind']=wind_le.fit_transform(df_new['wind'])
df_new['play']=play_le.fit_transform(df_new['play'])
```

```python
df
```

```python
df_new
```

```python
## split our data into X and y
```

```python
X = df_new.drop('play', axis =1)
```

```python
X
```

```python
y = df_new['play']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
# split out data into training and testing set

Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=0.20)
```

```python
Xtrain.shape
```

```python
Xtest.shape
```

```python
from sklearn.linear_model import LinearRegression  ### last session
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
## build the model classification-> logistic regression

#initiate the object
log_model = LogisticRegression()
```

```python
Xtrain
```

```python
ytrain
```

```python
## train the model using train data

log_model.fit(Xtrain, ytrain)
```

```python
ytrain_pred=log_model.predict(Xtrain)
```

```python

from sklearn.metrics import r2_score ## last session for regression 
from sklearn.metrics import accuracy_score ## for classification
```

```python

## checking accuracy of the model for training set
accuracy_score(ytrain, ytrain_pred)

## training score: 81%
```

```python
### checking accuracy for testing set

ytest_pred=log_model.predict(Xtest)
```
